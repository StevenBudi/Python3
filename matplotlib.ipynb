```python
df["Order Date"] = df["Order Date"].apply(lambda x : str(x))
transaksi = list(filter(lambda x : str(x[:4]) == "2015", df["Order Date"]))
```

```python
months = {}
for i in transaksi:
    if i[5:7] not in months:
        months[i[5:7]] = 1
    else :
        months[i[5:7]] += 1
```

```python
import collections
```

```python
months = collections.OrderedDict(sorted(months.items()))
```

```python
bulan = [x for x, _ in months.items()]
penjualan = [v for _, v in months.items()]
plt.bar(bulan, penjualan)
plt.title("Penjualan Tahun 2015")
plt.xlabel("Bulan")
plt.ylabel("Jumlah Penjualan")
plt.show()
```

```python
df = pd.read_csv("weight-height.csv")
```

```python
df.head(3)
```

```python
tinggi = list(df["Height"])
berat = list(df["Weight"])
plt.scatter(tinggi, berat)
plt.xlabel("Tinggi(cm)")
plt.ylabel("Berat(pound)")
plt.show()
```

```python
rataBerat = np.mean(berat)
rataTinggi = np.mean(tinggi)
stdBerat = np.std(berat)
stdTinggi = np.std(tinggi)

dataClean = df[(np.abs((df["Weight"] - rataBerat)/stdBerat)<= 3) & (np.abs((df["Height"] - rataTinggi)/stdTinggi) <=3)]
```

```python
berat = dataClean["Weight"]
tinggi = dataClean["Height"]
plt.scatter(berat, tinggi)
plt.show()
```
